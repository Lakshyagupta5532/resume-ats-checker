```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

resume_text = extract_text_from_pdf('sample_resume.pdf')
print(resume_text)
```

```python
job_description = """
We are looking for a Data Analyst with strong skills in Python, SQL, 
Excel, and data visualization tools like Tableau or Power BI. 
Experience with machine learning, statistics, and pandas is a plus. 
Strong communication skills and ability to work with large datasets required.
"""

print(job_description)
```

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def clean_text(text):
    text = text.lower()                          # lowercase
    text = re.sub(r'[^a-z\s]', ' ', text)         # sirf letters rakho, numbers/symbols hatao
    words = text.split()                          # words mein todo
    words = [w for w in words if w not in ENGLISH_STOP_WORDS]  # stopwords hatao
    return " ".join(words)

cleaned_resume = clean_text(resume_text)
cleaned_jd = clean_text(job_description)

print(cleaned_resume)
print("---")
print(cleaned_jd)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Dono texts ko ek list mein daalo
documents = [cleaned_resume, cleaned_jd]

# TF-IDF Vectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

# Similarity calculate karo
similarity_score = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])
match_percentage = similarity_score[0][0] * 100

print(f"Match Score: {match_percentage:.2f}%")
```

```python
# Resume aur JD ke unique words nikalo
resume_words = set(cleaned_resume.split())
jd_words = set(cleaned_jd.split())

# JD mein jo words hain but resume mein nahi hain
missing_skills = jd_words - resume_words

print("Missing Skills/Keywords:")
print(missing_skills)
```

```python
# Common skills list (tu apne field ke hisab se badha sakta hai)
skills_list = ['python', 'sql', 'excel', 'tableau', 'power', 'bi', 'machine', 
               'learning', 'statistics', 'pandas', 'numpy', 'visualization',
               'communication', 'analysis', 'data', 'java', 'javascript']

def extract_skills(text, skills_list):
    words = set(text.split())
    found_skills = words.intersection(skills_list)
    return found_skills

resume_skills = extract_skills(cleaned_resume, skills_list)
jd_skills = extract_skills(cleaned_jd, skills_list)

missing_skills = jd_skills - resume_skills

print("Skills in JD:", jd_skills)
print("Skills in Resume:", resume_skills)
print("Missing Skills:", missing_skills)
```

```python
import yake

def extract_keywords(text, num_keywords=15):
    kw_extractor = yake.KeywordExtractor(lan="en", n=2, top=num_keywords)
    keywords = kw_extractor.extract_keywords(text)
    return [kw[0].lower() for kw in keywords]

jd_keywords = extract_keywords(job_description)
print("JD Keywords:", jd_keywords)
```

```python
def check_keywords_in_resume(keywords, resume_text):
    resume_lower = resume_text.lower()
    found = [kw for kw in keywords if kw in resume_lower]
    missing = [kw for kw in keywords if kw not in resume_lower]
    return found, missing

found_skills, missing_skills = check_keywords_in_resume(jd_keywords, resume_text.lower())
print("Found:", found_skills)
print("Missing:", missing_skills)
```

```python
match_percentage = (len(found_skills) / len(jd_keywords)) * 100
print(f"Match Score: {match_percentage:.2f}%")
```

```python
def get_required_skills(job_title):
    prompt = f"""List the top 15 most important technical skills, tools, and qualifications 
    required for a {job_title} role. Return ONLY a list separated by the pipe symbol (|), 
    nothing else. No explanations, no numbering, no parentheses.
    Example format: python|sql|machine learning|aws"""
    
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )
    skills_text = response.text
    skills = [s.strip().lower() for s in skills_text.split("|")]
    return skills

job_title = "AIML Engineer"
required_skills = get_required_skills(job_title)
print(required_skills)
```

```python
def check_skills_in_resume(skills, resume_text):
    resume_lower = resume_text.lower()
    found = []
    missing = []
    
    for skill in skills:
        # agar skill mein / ya & hai, sub-parts mein check karo
        parts = skill.replace('&', '/').split('/')
        if any(part.strip() in resume_lower for part in parts):
            found.append(skill)
        else:
            missing.append(skill)
    
    return found, missing

found_skills, missing_skills = check_skills_in_resume(required_skills, resume_text)
match_percentage = (len(found_skills) / len(required_skills)) * 100

print(f"Match Score: {match_percentage:.2f}%")
print("Found:", found_skills)
print("Missing:", missing_skills)
```
